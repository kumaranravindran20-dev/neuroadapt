```python
from neuroadapt import AdaptiveModel

model = AdaptiveModel(learning_rate=0.05)

data_stream = [
    (1, 2),
    (2, 4),
    (3, 6),
    (4, 8),
]

for x, y in data_stream:
    prediction = model.predict(x)
    model.adapt(x, y)
    print(f"x={x}, prediction={prediction:.2f}, target={y}")
```
