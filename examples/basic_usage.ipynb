```python
from neuroadapt import AdaptiveModel
import random

model = AdaptiveModel()

errors = []

for i in range(100):
    x = random.uniform(0, 10)
    y = 2 * x   # true relationship

    pred = model.predict(x)
    error = abs(y - pred)
    errors.append(error)

    model.adapt(x, y)

print("Initial error:", errors[0])
print("Final error  :", errors[-1])
```

```python
from neuroadapt import AdaptiveLSTMModel

model = AdaptiveLSTMModel(input_size=1, window_size=3)

data = [1, 2, 3, 4, 5, 6]

for i in range(3, len(data)):
    window = [[data[i-3]], [data[i-2]], [data[i-1]]]
    target = data[i]

    before = model.predict(window)
    model.adapt(window, target)
    after = model.predict(window)

print("AdaptiveLSTMModel test passed")
```
